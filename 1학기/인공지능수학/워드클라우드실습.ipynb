```python
from wordcloud import WordCloud     # 워드클라우드 생성 라이브러리
import matplotlib.pyplot as plt     # 워드클라우드 이미지를 화면에 출력하기 위한 라이브러리
from collections import Counter
import re 

-- re는 resgular expression
```

**텍스트 전처리 함수**

```python
def preprocess_text(text: str) -> list[str]:
    """
    텍스트 문장 전처리:
    - 영문/숫자/한글만 남김
    - 2글자 이상 단어만 사용
    시험에 나온다
    """
    text = re.sub(r"[^0-9A-Za-z가-힣\s]", " ", text)
    words = text.split()
    words = [w.lower() for w in words if len(w) >= 2]
    return words
```

**워드 클라우드 생성 함수**

```python
def make_wordcloud(
    input_sentence: str,
    font_path: str | None = None
) -> None:
    """
    텍스트 문장에서 단어 빈도수를 계산해 워드클라우드를 생성한다.
    """
    words = preprocess_text(input_sentence)


    # 불용어(stop word) 리스트
    stopwords = {
        "am", "is", "was", "are", "were", "on", "such", "as", "it", "on", "in",
        "under", "only", "such", "we", "our", "us", "will", "have", "among", "has",
        "into", "former", "through", "out",  ",", ".", ":", ";", "(", ")"
        "a", "an", "the", "this", "those", "that", "and", "or", "not", "for",
        "with", "to", "so", "thus", "he", "she", "his", "her", "of", "which",
        "who", "whose", "what", "where", "when", "how", "why"
    }

    words = [w for w in words if w not in stopwords]


    # 단어 빈도수 계산
    freq = Counter(words)


    # 워드 클라우드 생성
    wc = WordCloud(
        width=1000,
        height=600,
        background_color="white",
        font_path=font_path,   # 한글이면 폰트 경로 지정 필요
        max_words=200
    ).generate_from_frequencies(freq)


    # 워드 클라우드 이미지 파일 저장
    # (output_path에 워드클라우드 이미지 파일이 저장될 경로와 파일 이름 기입)
    # wc.to_file(output_path)


    # 워드 클라우드 이미지를 화면에 표시
    plt.figure(figsize=(10, 6))
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.tight_layout()
    plt.show()
```
